```python
sns.heatmap(s.corr(),annot=True,fmt=".2f")
plt.title("Correlation of advertising mediums with sales")
plt.show()
```

```python
x=s[["TV"]]
y=s[["Sales"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
ML=LinearRegression()
ML.fit(x_train,y_train)
x_train_prediction=ML.predict(x_train)
print(x_train_prediction)
print(y_train)
print(ML.coef_)
print(ML.intercept_)
x_test_prediction=ML.predict(x_test)
print(x_test_prediction)
print(y_test)
print(ML.coef_)
print(ML.intercept_)
x_test_prediction=ML.predict(x_test)
print(x_test_prediction)
print(y_test)
```

```python
plt.figure(figsize=[7,5])
plt.scatter(y_train,x_train_prediction)
plt.xlabel("Original values")
plt.ylabel("Predicted values")
plt.title("Original values VS Predicted values")
plt.show()
```

```python
plt.figure(figsize=[7,5])
plt.scatter(y_test,x_test_prediction)
plt.xlabel("Original values")
plt.ylabel("Predicted values")
plt.title("Original values VS Predicted values")
plt.show()
```
